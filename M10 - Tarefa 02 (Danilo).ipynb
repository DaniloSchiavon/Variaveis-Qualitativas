````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('G:\EBAC\Provas\Modulo 10\previsao_de_renda.csv')
````

````python
# Gráfico posse de imóvel
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="posse_de_imovel", y="renda", data=df, dodge=True, ci=95)
ax.set_title('Renda Média por Posse de Imóvel')
ax.set_xlabel('Posse de Imóvel')
ax.set_ylabel('Renda')
plt.show()

# Gráfico posse de veículo
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="posse_de_veiculo", y="renda", data=df, dodge=True, ci=95)
ax.set_title('Renda Média por Posse de Veículo')
ax.set_xlabel('Posse de Veículo')
ax.set_ylabel('Renda')
plt.show()
````

````python
# Estatísticas descritivas para posse de imóvel
imovel_stats = df.groupby('posse_de_imovel')['renda'].describe()
print("Estatísticas descritivas para posse de imóvel:")
print(imovel_stats)

# Estatísticas descritivas para posse de veículo
veiculo_stats = df.groupby('posse_de_veiculo')['renda'].describe()
print("Estatísticas descritivas para posse de veículo:")
print(veiculo_stats)
````

````python
# Gráfico de ponto para posse de imóvel com hue para sexo
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue='sexo', data=df, dodge=True, ci=95)
ax.set_title('Renda Média por Posse de Imóvel e Sexo')
ax.set_xlabel('Posse de Imóvel')
ax.set_ylabel('Renda')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.30), ncol=3)
plt.show()

# Gráfico de ponto para posse de veículo com hue para sexo
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue='sexo', data=df, dodge=True, ci=95)
ax.set_title('Renda Média por Posse de Veículo e Sexo')
ax.set_xlabel('Posse de Veículo')
ax.set_ylabel('Renda')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.30), ncol=3)
plt.show()
````

### Insights:

1) Nos primeiros gráficos, é possível notar que as pessoas que possuem e veículo ou possuem imóvel ganham mais do que as pessoas que não possuem, dentro do intervalo de confiança de 95%.
2) Dentro da base de dados, no segundo gráfico, é notável a diferença salarial entre homens e mulheres, porém o mesmo padrão do item 01 se repete nesse item também.
3) No terceiro gráfico, mesmo separando o sexo, o padrão entre renda e posse de imóvel se mantém. Ou seja, independente do sexo, uma pessoa que possui uma renda maior tem maiores chances de possuir uma imóvel.
4) No quarto gráfico, é possível notar que as amostras são inversamente proporcionais, ou seja, quando juntamos as variáveis 'Renda Média', 'Posse de Veículo' e 'Sexo', para o sexo masculino, conforme a renda diminui, mesmo que assimétrico, a média salarial é menor dos homens que possuem veículo do que dos homens que não possuem veículos, e, no caso do sexo feminino, é justamente o oposto, ou seja, conforme a renda aumenta, é mais provável da pessoa possuir um veículo.
5) Ainda sobre o quarto gráfico, quando separamos os sexos, é possível ver que os homens, mesmo com a renda inferior em relação a média dos outros homens, possuem mais chances de terem um veículo, oque não acontece quando juntamos o sexo e analisamos simplesmente a 'renda x posse de veículo'.
